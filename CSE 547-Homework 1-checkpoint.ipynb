```python
# CSE 547-02 Deep Learning Homework 1

#Loading in the data
from tensorflow.keras.datasets import mnist
import pandas as pd
import numpy as np



#splitting the data into training and testing section
(train_images,train_labels), (test_images, test_labels) = mnist.load_data()



#preparing the training and testing images for a fully connected network that expect 1-D input
train_images = train_images.reshape((60000, 28*28)) 
train_images = train_images.astype('float32')/255

#test_images = test_images.reshape((10000, 28*28))
#test_images = test_images.astype('float32')/255



print(f"Executed successfully") 
```

```python
#importing the libraries
from tensorflow.keras import models
from tensorflow.keras import layers


#Network configuration for a fully connected network/a sequential network
network = models.Sequential()

#adding a hidden layer with 512 nodels in the first layer with 784 features
network.add(layers.Dense(512, activation='relu', input_shape=(784,)))

#the output has 10 classes 
network.add(layers.Dense(10, activation = 'softmax'))

# Compile the model
network.compile(
    optimizer='rmsprop',
    loss='categorical_crossentropy', #use this because data is one-hot encoded
    metrics=['accuracy']
)

# Summary of the model
network.summary()
```

```python
#training models each with 4 folds cv
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import KFold
import tensorflow as tf


kf = KFold(
    n_splits=4,
    shuffle=True,
    random_state=42
)

results = []

#one layer=1 node
#2 layers = 12,4 nodes
#3 layers = 32 nodes, 32 nodes, and 16 nodes

for layers_config in [[1], [12, 4], [32, 32, 16]]:
    #defining learning rate each number of layer will train with
    for lr in [0.1, 0.001, 0.00001]:

        fold_accuracies = []

        for fold, (train_idx, val_idx) in enumerate(kf.split(train_images)):
            X_train, X_val = train_images[train_idx], train_images[val_idx]
            y_train, y_val = train_labels[train_idx], train_labels[val_idx]

            model = build_model(layers_config, lr)

            history = model.fit(
                X_train, y_train,
                epochs=20,
                batch_size=128,
                validation_data=(X_val, y_val),
                verbose=0
            )

            # Plot accuracies
            plt.figure()
            plt.plot(history.history['accuracy'], label='train')
            plt.plot(history.history['val_accuracy'], label='val')
            plt.title(f"Layers={layers_config}, LR={lr}, Fold={fold+1}")
            plt.xlabel("Epoch")
            plt.ylabel("Accuracy")
            plt.legend()
            plt.show()

            fold_accuracies.append(max(history.history['val_accuracy']))

        avg_acc = np.mean(fold_accuracies)

        results.append({
            "layers": layers_config,
            "learning_rate": lr,
            "avg_val_accuracy": avg_acc
        })
```
